#Q5 Meaningful Vizualisation

# Line Graph Of Flower Species Iris Virginica between Sepal width and Petal Length

```python
data=data.head()
df= pd.DataFrame(data, columns=["SW","PL","Iris-virginica"])
df.plot(x="SW",y=["PL","Iris-virginica"], kind= "line", figsize=(9,9))
plt.xlabel("Sepal Width ")
plt.ylabel("Petal Length")
plt.show()
```

#Line Graph Of Flower Species Iris Setosa between Petal width and Petal Length

```python
data=data.head()
df= pd.DataFrame(data, columns=["PW","PL","Iris-setosa"])
df.plot(x="PW",y=["PL","Iris-setosa"], kind= "line", figsize=(9,9))
plt.xlabel("Petal Width ")
plt.ylabel("Petal Length")
plt.show()
```

#Line Graph Of Flower Species Iris Versicolor between Sepal width and Sepal Length

```python
data=data.head()
df= pd.DataFrame(data, columns=["SW","SL","Iris-versicolor"])
df.plot(x="SW",y=["SL","Iris-versicolor"], kind= "line", figsize=(9,9))
plt.xlabel("Sepal Width ")
plt.ylabel("Sepal Length")
plt.show()
```
